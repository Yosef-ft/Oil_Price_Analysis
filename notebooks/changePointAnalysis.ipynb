- **2008-04-01: Global financial crisis begins affecting demand:** This event would lead to a price decrease due to reduced demand.

- **2009-04-01: Early signs of economic recovery:** This event could lead to a price increase as demand starts to recover.
- **2015-01-02: Oil price crash due to oversupply:** This event would lead to a price decrease.
- **2015-10-01: Global economic concerns affect demand:** This event would lead to a price decrease.

- **2020-04-01: COVID-19 pandemic leads to historic price drop:** This event would lead to a significant price decrease due to reduced demand and supply chain disruptions.
- **2020-07-01: Recovery in demand as economies reopen:** This event would lead to a price increase as demand recovers.
- **2021-01-04: Vaccine rollout boosts global economic outlook:** This event would lead to a price increase as economic activity and demand for oil increase.
